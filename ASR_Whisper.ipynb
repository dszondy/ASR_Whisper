## Transcription

```python
from transformers import pipeline

transcriber = pipeline("automatic-speech-recognition", model='openai/whisper-large-v2', chunk_length_s=30, generate_kwargs = {"language":"<|en|>","task": "transcribe"},)
```

```python
import subprocess

## My phone saves audio as aac by default but it did not work with whisper
def convert_aac_to_mp3(aac_file_path, mp3_file_path):
    command = [
        'ffmpeg',
        '-i', aac_file_path,  # Input file
        '-acodec', 'libmp3lame',  # Use MP3 encoder
        mp3_file_path  # Output file
    ]
    subprocess.run(command, check=True)

aac_file = input_file
mp3_file = 'temp.mp3'
convert_aac_to_mp3(aac_file, mp3_file)
```

```python
transcription = transcriber(mp3_file)
```

```python
with open("Transcript.txt", 'a') as file:
    file.write(transcription["text"])
```
